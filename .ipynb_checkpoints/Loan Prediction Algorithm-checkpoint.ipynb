# Basic details of project
###### Business objective:
For automating Loan approval process platform to reduce the time of processing. 

So we should not miss any loan which is eligible, but we have not approved it through alogarithm. That is high recall ratio for this model is required.

Hypothesis:
1) Null Hypothesis: There is no relationship between loan approval process and Individual parameters of the model.
2) Alternatie Hypothesis: There is a relationship existing.


#### Problem Statement: 
                       Automation of loan eligibility process by using provided data sets. Considered objective atleast one healthy loan payer should not miss the loan. since thats a business loss to the company.
#### Solution Process:
	1-> Statistical understanding of features by doing Chi Square test. Null Hypothesis is failed for all the categorical variables.
	2-> Doing ANOVA with target variable for all the continuous variables. Null Hypothesis is failed for Credit History variables.
	3-> Data Preparation 1-> Have prepared data using mean as replacement for continuous variables and most frequent for categorical variables.
						 2-> Used pandas getdummies method for converting categorical variables to numerical numbers to run ML algorithms.
						 3-> As shown in the code I have used Decision Tree Classification algorithm for building predictive model. 
						 	 And used GridSearchedCV to tune model parameters to get a best score 70% accuracy. Through this model I figured out 
						 	 important features to predict the target parameter Loan Status are Credit History,Property_Area_Rural,LoanAmount,Married_Yes,Total_Inc.
						 4-> Picked the dataset which are not blank for Credit History,Property_Area,LoanAmount,Married.
	4-> Model Building   1->Reiterated 3.1,3.2 on 3.4 dataset (row count 541). Since Train dataset to be clean and good with real information for better model building. 
						 2->Used RandomForest Classifier ensemble algorithm and used GridSearchCV for hyper tuning of model achieved below scores
						 Accuracy of the model- 77.7%
						 
						 
I have used the above model to test sampled data from test dataset (109 loans IDs) and the scores are below:
						 Area under the curve- 85.7%
						 F1_Score of model 93.75%
						 Precision Score 89.29%
						 Recall Score 98.68%
						 Confusion metrics -
						 [[24  9]
 						  [ 1 75]]
						 In the above confusion matrix - False positive is only in 100. And thats the same we expect from model.
						 One healthy customer should not miss loan, as its a threat to customers.  
						 The model is not overfit or under fit.
#### Solution Description: 
This model is built using ensemble algorithm called Random Forest Classifier ML Algorithm. With this properties
					Class weight is "balanced", number of trees built are 25 with maximum depth 10 and min_samples_split is 2% of 541.
					This model gave accuracy as 77% and Area Under the curve as 85.7%. Crucial point is the False positive of this model
					are one in hundred. Which gives an understanding among one in hundred an application of healthy applicant get missed.

# ChiSquare Test
This test has been performed to check the influence of one categorical parameter to other categorical parameters (i.e. Loan Status).
H0-> Loan_status group is equally distributed across other categorical parameter
H1->Loan_status group is not equally distributed across other categorical parameter

```python
lst_cate=data_train.select_dtypes(include='object').columns
for t in lst_cate[1:-1]:
    temp_dt= data_train.reset_index()[['Loan_ID','Loan_Status',t]]
    temp_dt1=pd.pivot_table(temp_dt,values='Loan_ID',index=t,columns='Loan_Status',aggfunc='count')
    #print(temp_dt1)
    print('Chisquare results for {} and loan status {:0.7f}, total obs {}'.format(t,stats.chisquare(temp_dt1.values[:][0],temp_dt1.values[:][1])[1],temp_dt1.values.sum()))
    
```

# ANOVA test
Below I have done Anova to check the mean of multiple numerical parameters for Loan approved or Not has any significant difference.
H0-> There is not significant difference between means of two groups
H1-> There is significant difference

```python
#help(stats.chisquare)

for t in list_num:
    temp_dt1=data_train[data_train['Loan_Status']=='Y'][t].dropna()
    temp_dt2=data_train[data_train['Loan_Status']=='N'][t].dropna()
    p=stats.f_oneway(temp_dt1.values,temp_dt2.values)
    print('ANOVA results for {} and loan status {:0.7f}'.format(t,p[1]))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(class_weight='balanced',max_depth=20,max_features=8,min_samples_leaf=0.1,random_state=10)
X_train,X_test,Y_train,Y_test=train_test_split(data_train_filt_dummies2_trans[:,:-1],data_train_filt_dummies2_trans[:,-1],test_size=0.2,random_state=11)
dtc.fit(X_train,Y_train)
Y_pred=dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,roc_auc_score,classification_report,f1_score,precision_score,recall_score,precision_recall_curve,roc_curve
print(confusion_matrix(Y_test,Y_pred))
print('ROC_AUC_Score {:.4f}'.format(roc_auc_score(Y_test,Y_pred)))
print('F1_Score of model {:.4f}'.format(f1_score(Y_test,Y_pred)))
print("Precision Score {:.4f}".format(precision_score(Y_test,Y_pred)))
print("Recall Score {:.4f}".format(recall_score(Y_test,Y_pred)))
```

```python
classification_report(Y_test,Y_pred)
```

```python
fpr,tpr,thresholds=roc_curve(Y_test,Y_pred)
def plot_roc_curve(fpr,tpr,label=None):
    plt.plot(fpr,tpr,linewidth=2,label=label)
    plt.plot([0,1],[0,1],'k--')
    plt.axis([0,1,0,1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
                                                                                                                                                                                                                                    
plot_roc_curve(fpr,tpr)                                                                                                                                                                                                                                    
plt.show()            
```
